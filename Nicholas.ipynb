```python
#FUNCTIONS
def workers(name,age,salary):
    print(f'My name is {name}, I am {age} years old. I earn GHC{salary} per month')
workers("Nicholas",59,2000)
```

```python
def my_children(*names):
    for name in names:
        print(name)
my_children('Theresah','Anna','Samuel','Nicholas','Gabriel')
```

```python
def my_children(**names):
    for key, value in names.items():
        print(key,value)
my_children(first='Theresah',second='Anna',third='Samuel',forth='Nicholas',fifth='Gabriel')
```

```python
#STUDENT DATABASE
import sqlite3
conn=sqlite3.connect('student_database.db')
cursor=conn.cursor()

#STUDENT DATA TABLE CREATION
cursor.execute('''
    CREATE TABLE IF NOT EXISTS  students(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        firstname TEXT NOT NULL,
        lastname TEXT NOT NULL,
        email TEXT UNIQUE NOT NULL,
        phone TEXT NOT NULL)''')
conn.commit()
conn.close()
print('Student Database successfully created')
```

```python
#INSERTING STUDENT DATA
conn=sqlite3.connect('student_Database.db')
cursor=conn.cursor()
cursor.execute('''
INSERT INTO students (firstname,lastname,email,phone)
  VALUES (?,?,?,?)''',(firstname,lastname,email,phone))
conn.commit()
conn.close()
```
