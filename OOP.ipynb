```python
#Object Oriented Programming Trials.
#Creating a class of cars.
class cars:
    def __init__(self,brand,model,color):
        self.brand=brand
        self.model=model
        self.color=color

    def moving(self):
        print(f'This {self.brand}  {self.model} moves fast')
Toyota=cars('Toyota','Highlander','white')
Hundai=cars('Hundai','sonator','Blue')
Hunda=cars('Hunda','CR-V','Black')
Toyota.moving()
Hundai.moving()
Hunda.moving()
```
