```python
#TRYING MY HANDS ON BRUTE FORCE ALGORITHM
#BY THE WAY, BRUTE FORCE ALGORITHM IS AN ALGORITHM, THAT INVOLVES TRYING ALL POSSIBLE MEANS 
#IN SOLVING A PROBLEM TO FIND THE BEST ONE OR OPTION
#SO BELOW IS A CODE THAT SHOWS AN EXAMPLE OF A BRUTE FORCE ALGORITHM
def brute_force(arr,target):
    #iterating through the list.
    for i in range(len(arr)):
        if arr[i]==target:
            return i
        
    return "Not found"
numbers=[3,5,7,9,10,38]
print(brute_force(numbers,38))
```
