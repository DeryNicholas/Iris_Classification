```python
#CLASSES
class car:
        def turn_on(self):
                print("You turn on this car")
                return self
        def brake(self):
                print("You step on the brake")
                return self
        def drive(self):
                print("start to drive")
                return self
        def stop(self):
                print("The engine has been stopped")
                return self
car=car()
car.brake()
car.turn_on()
car.drive()
car.brake()
car.stop()
```

```python
#Walrus operator :=
foods=list()
while True:
    food=input("What food do you like most?")
    if food=="quit":
        break
    foods.append(food)
  
```

```python
#Using Walrus operator for the function
foods=list()
while food:=input("What food do you like most?")!="quit":
    foods.append(food)        
```

```python
#HIGHER ORDER FUNCTIONS
def loud(text):
    return text.upper()

def quiet(text):
    return text.lower()

def nick(func):
    text=func("hello Nick")
    print(text)

nick(loud)
nick(quiet)
```

```python
#SORT FUNCTION
workers=["John","Akwasi","Maalam","Bright","George"]
sorted_workers=sorted(workers)
for i in sorted_workers:
    print(i)
```

```python
students=[("Nicholas","M",40),("John","M",20),("Mary","F",30),
          ("Carol","F",10)]
grade=lambda grades:grades[1]
students.sort(key=grade)
for i in students:
    print(i)
```

```python
#MAP FUNCTION
store=[("Trouser",300),
       ("Shirt",200),
       ("Book",100),
       ("Pen",500),
       ("Erazer",400)]

to_euros=lambda data:(data[0],data[1]*15.00)
price_in_euros=map(to_euros,store)
for i in price_in_euros:
    print(i)
```

```python
#FILTER FUNCTION
classmates=[("Osaah",40),
            ("Bright",60),
            ("Sunshine",20),
            ("Carol",10),
            ("Aaron",40)]
age=lambda data:data[1]>30
mates_older_than_30=list(filter(age,classmates))
for i in mates_older_than_30:
    print(i)
```
