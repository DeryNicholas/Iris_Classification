```python

import sqlite3
import tkinter as tk
from tkinter import messagebox

window = tk.Tk()

# --- CREATING LABELS AND ENTRIES ---
Yourname_label = tk.Label(window, text='Yourname')
Yourname_label.grid(row=0, column=0)
Yourname_Entry = tk.Entry(window)
Yourname_Entry.grid(row=0, column=1)

Fathersname_label = tk.Label(window, text='Fathersname')
Fathersname_label.grid(row=1, column=0)
Fathersname_Entry = tk.Entry(window)
Fathersname_Entry.grid(row=1, column=1)

Mothersname_label = tk.Label(window, text='Mothersname')
Mothersname_label.grid(row=2, column=0)
Mothersname_Entry = tk.Entry(window)
Mothersname_Entry.grid(row=2, column=1)

# --- DATABASE FUNCTIONS ---
def create_database():
    try:
        conn = sqlite3.connect('Family_database.db')
        cursor = conn.cursor()
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS Family (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                Yourname TEXT NOT NULL,
                Fathersname TEXT NOT NULL,
                Mothersname TEXT NOT NULL
            )
        ''')
        conn.commit()
        conn.close()
        print('Database successfully created')
    except sqlite3.Error as e:
        messagebox.showerror('Database error', str(e))  # Show the error message

def insert_data():
    Yourname = Yourname_Entry.get()
    Fathersname = Fathersname_Entry.get()
    Mothersname = Mothersname_Entry.get()

    if not Yourname or not Fathersname or not Mothersname:
        messagebox.showerror('Error', 'All fields must be filled')
        return

    try:
        conn = sqlite3.connect('Family_database.db')
        cursor = conn.cursor()
        cursor.execute('''
            INSERT INTO Family(Yourname, Fathersname, Mothersname)
            VALUES (?, ?, ?)
        ''', (Yourname, Fathersname, Mothersname))
        conn.commit()
        conn.close()

        messagebox.showinfo('Success', 'Data successfully added')
        clear_data()
    except sqlite3.Error as e:
        messagebox.showerror('Error', 'There is an error while inserting your data: ' + str(e))

def clear_data():
    Yourname_Entry.delete(0, tk.END)
    Fathersname_Entry.delete(0, tk.END)
    Mothersname_Entry.delete(0, tk.END)

# --- BUTTONS ---
summit_button = tk.Button(window, text='Submit', command=insert_data)
summit_button.grid(row=3, column=0)

# --- INITIALIZATION ---

create_database() #Create the database before the app runs.
#Always create the database before attempting to use it.
window.mainloop()




```
