```python
class student:
    def __init__(self, name,student_id,gpa):
        self.name=name
        self.student_id=student_id
        self.gpa=gpa
    def display_info(self):
        print(f'Name: {self.name}')
        print(f'student_id :{self.student_id}')
        print(f'gpa :{self.gpa}')
    
    def __str__(self):
        return f'{self.name}|{self.student_id}|{self.gpa}'
student_db=[]
def create_student():
    name=input('Enter student name:')
    student_id=int(input('Enter student ID'))
    gpa=float(input('Enter student GPA'))

    new_student=student(name,student_id,gpa)
    student_db.append(new_student)

    print(f'{name} has been sucessfully added to the system')
def read_student():
    id=int(input('Enter student ID'))
    found=False
    for student in student_db:
        if student.student_id==id:
            print('\nstudent found')
            student.display_info()
            found=True
            break
    if not found:
        print('student not found\n')
def update():
    id=int(input('Enter student ID to update credentials:'))
    for student in student_db:
        if student.student_db==id:
            print("/nCurrent credentials:")
            student.display_info()
            print('What would you like to change')
            print('1.Name')
            print('2.GPA')
            choice=int(input('Select 1/2'))
            if choice ==1:
                new_name=input('Enter new name:')
                student.name=new_name
            elif choice==2:
                new_gpa=input('Enter new GPA')
                student.gpa=new_gpa
            else:
                print('Invalid option')
                return
    
create_student()
read_student()
print('All the students in the system')
for student in student_db:
    print(student)


```
